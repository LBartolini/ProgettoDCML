# Training

```python
from sklearn.model_selection import train_test_split

from sklearn.ensemble import RandomForestClassifier, StackingClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB

from sklearn.metrics import accuracy_score, confusion_matrix

from keras.models import Sequential
from keras.layers import Dropout, LSTM, Dense, Flatten
from keras.optimizers import Adam
from keras.losses import BinaryCrossentropy
```

### Train Test Split

```python
X_train_shuffled, X_test_shuffled, y_train_shuffled, y_test_shuffled = train_test_split(X_std, y, test_size=0.3, shuffle=True) # split shuffling
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_std, y, test_size=0.2, shuffle=False) # split without shuffling (used in RNN)
```

### RandomForestClassifier

```python
random_forest = RandomForestClassifier(n_estimators=11, n_jobs=-1)
random_forest.fit(X_train_shuffled, y_train_shuffled)
predicted_proba = random_forest.predict_proba(X_test_shuffled)
predicted = random_forest.predict(X_test_shuffled)
print("Accuracy: ", accuracy_score(y_test_shuffled, predicted))
print("Confusion Matrix: \n", confusion_matrix(y_test_shuffled, predicted))
```

```python
from joblib import dump
dump(random_forest, root_path+'random_forest.bin', compress=3) # save the RandomForestClassifier (best model tested)
```











### Stacking
















```python
stacking_classifier = StackingClassifier(estimators=[('lda', LinearDiscriminantAnalysis()),
                                       ('nb', GaussianNB()),
                                       ('dtc', DecisionTreeClassifier())],
                            final_estimator=RandomForestClassifier(n_estimators=7))
stacking_classifier.fit(X_train_shuffled, y_train_shuffled)
predicted_proba = stacking_classifier.predict_proba(X_test_shuffled)
predicted = stacking_classifier.predict(X_test_shuffled)
print("Accuracy: ", accuracy_score(y_test_shuffled, predicted))
print("Confusion Matrix: \n", confusion_matrix(y_test_shuffled, predicted))
```

### Neural Network

```python
NUM_EPOCHS = 20
BATCH_SIZE = 15 # numbers of samples used for each step

TIME_STEPS = 4 # individual monitored data (row) in each sample
OVERLAP = 2 # used to make two samples overlap
N_FEATURES = X_std.shape[1]

INIT_LR = 0.05

model = Sequential([
    LSTM(64, input_shape=(TIME_STEPS, N_FEATURES)),
    Dropout(0.3),
    Dense(32, activation='relu'),
    Dropout(0.3),
    Dense(8, activation='relu'),
    Dense(1, activation='sigmoid'),
    Flatten()
])
model.build()
model.compile(
    loss=BinaryCrossentropy(),
    optimizer=Adam(learning_rate=INIT_LR, decay=INIT_LR / (NUM_EPOCHS * 0.5)),
    metrics=['accuracy']
)
model.summary()
```

```python
X_train = np.array(X_train)
X_test = np.array(X_test)
y_train = np.array(y_train)
y_test = np.array(y_test)

X_samples_train = []
y_samples_train = []

X_samples_test = []
y_samples_test = []

# building training set
for i in range(0, X_train.shape[0], int(TIME_STEPS/OVERLAP)):
  X_sample_train = X_train[i:i+TIME_STEPS]
  y_sample_train = y_train[i:i+TIME_STEPS]
  if X_sample_train.shape[0] == TIME_STEPS and y_sample_train.shape[0] == TIME_STEPS: # to avoid using partial samples
    X_samples_train.append(X_sample_train)
    y_samples_train.append(y_sample_train)

# building test set
for i in range(0, X_test.shape[0], int(TIME_STEPS/OVERLAP)):
  X_sample_test = X_test[i:i+TIME_STEPS]
  y_sample_test = y_test[i:i+TIME_STEPS]
  if X_sample_test.shape[0] == TIME_STEPS and y_sample_test.shape[0] == TIME_STEPS: # to avoid using partial samples
    X_samples_test.append(X_sample_test)
    y_samples_test.append(y_sample_test)
```

```python
X_samples_train = np.array(X_samples_train)
y_samples_train = np.array(y_samples_train)

X_samples_test = np.array(X_samples_test)
y_samples_test = np.array(y_samples_test)
```

```python
model.fit(X_samples_train, y_samples_train,
          validation_data=(X_samples_test, y_samples_test),
          batch_size=BATCH_SIZE, epochs=NUM_EPOCHS)
```
